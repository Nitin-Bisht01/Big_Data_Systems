# Start Spark Session

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("AdvancedOps").getOrCreate()
```

# Read Dataset

```python
df = spark.read.option("header", "true").csv("/content/MYDATA.csv")
df.show()
```

#Filtering Operations
#1. Filter Rows by Conditions
# Single condition


```python
df.filter(df["age"] > 25).show()
```

# Multiple columns

```python
df.filter((df["age"] > 25) & (df["salary"] < 50000)).show()
```

#2. Logical Operators
# AND operator (&)

```python
df.filter((df["age"] > 25) & (df["department"] == "IT")).show()
```

# OR operator (|)

```python
df.filter((df["age"] > 40) | (df["salary"] > 80000)).show()
```

# NOT operator (~)

```python
df.filter(~(df["department"] == "HR")).show()
```

#3. Comparison Operators

```python
df.filter(df["salary"] == 50000).show() # Equal
df.filter(df["salary"] != 50000).show() # Not equal
df.filter(df["age"] > 30).show() # Greater than
df.filter(df["age"] >= 30).show() # Greater or equal
df.filter(df["age"] < 30).show() # Less than
df.filter(df["age"].isin([25, 30, 35])).show() # In list
```

#Aggregation & Grouping
#4. GroupBy Operations

# Group by department, aggregate salary

```python
df.groupBy("department").agg({"salary": "avg"}).show()
```

# Multiple aggregations

```python
from pyspark.sql.functions import avg, max, count
df.groupBy("department").agg(
avg("salary").alias("avg_salary"),
max("salary").alias("max_salary"),
count("*").alias("emp_count")
).show()
```

#5. Overall Aggregated Values

# Single aggregations across entire dataset

```python
df.agg({"salary": "sum", "age": "avg"}).show()
```

# Using functions

```python
from pyspark.sql.functions import sum, avg, min, max
df.agg(
sum("salary").alias("total_salary"),
avg("age").alias("avg_age"),
min("salary").alias("min_salary")
).show()
```

#Formatting & Sorting

6. format_number & Alias

```python
from pyspark.sql.functions import format_number, round
```

# Format numbers

```python
from pyspark.sql.functions import avg, max, round, format_number, col
df.groupBy("department").agg(
    round(avg(col("salary").cast("double")), 2).alias("avg_salary"),
    format_number(max(col("salary").cast("double")), 2).alias("max_salary_formatted")
).show()
```

#7. Sort Data with orderBy

# Single column ascending

```python
df.orderBy("salary").show()
```

# Multiple columns (descending salary, ascending age)

```python
df.orderBy(col("salary").desc(), col("age")).show()
```

# After groupby

```python
df.groupBy("department").agg(avg("salary").alias("avg_salary")) \
.orderBy(col("avg_salary").desc()).show()
```
