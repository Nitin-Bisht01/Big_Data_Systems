```python
#select column
df = spark.read.option("header", True).option("inferschema", True).csv("/content/sample_data/california_housing_test.csv")
df.select(['population','households','housing_median_age']).show()
```

```python
#adding column
df = df.withColumn("Revised housing_median_age", df['housing_median_age']+5)
df.show()
```

```python
#Dropping column
df_final_columns = df.drop('longitude','latitude','total_rooms','total_bedrooms','median_income','median_house_value','housing_median_age')
df_final_columns.show()
```

```python
#Renaming column
df_final = df_final_columns.withColumnRenamed("Revised housing_median_age", "Final_median_age")
df_final.show(3)
```
